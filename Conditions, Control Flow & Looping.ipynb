##### The Python break and continue Statements

```python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        break # Break Statement
    print(n)
print('Loop ended.')
```

```python
n = 5
while n > 0:
    n -= 1
    if n == 2:
        continue
    print(n)
print('Loop ended.')
```

```python
n = 10
while n > 0:
    n -= 1
    if n % 2 == 0:
        continue
    print(n)
print('Loop ended.')
```

##### The else Clause

```python
n = 5
while n > 0:
    n -= 1
    print(n)
else:
    print('Loop done.')
```

```python
n = 5
while n > 0:
    n -= 1
    print(n)
    if n == 2:
        break
else:
    print('Loop done.')
```

```python
temp = input("Ketikan temperatur yang ingin dikonversi, eg. 45F, 120C: ")
degree = int(temp[:-1])
i_convertion = temp[-1]
result = 0
if i_convertion == "C":
    result = int(round((9 * degree) / 5 + 32))
elif i_convertion == "F":
    result = int(round((degree - 32) * 5 / 9))
else:
    print("Masukan input yang benar")

print("Temperaturnya adalah", result, "derajat")
```
